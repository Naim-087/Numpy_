# NumPy Searching (Simple + Complete Note)

## 🔷 Definition

Searching in NumPy means finding **values, indices, or positions** of elements in an array. It helps locate data efficiently using conditions or built-in functions.

---

## 🔷 1. Using `np.where()`

Returns indices where a condition is true.

### Example:

```python
import numpy as np
a = np.array([10, 20, 30, 40])

np.where(a == 30)
# Output: (array([2]),)
```

### Condition:

```python
np.where(a > 20)
# Output: (array([2, 3]),)
```

### 2D Example:

```python
a = np.array([[1,2,3],
              [4,5,6]])

np.where(a > 3)
# Output: (array([1,1,1]), array([0,1,2]))
```

👉 Returns (row_indices, column_indices)

---

## 🔷 2. Using `np.searchsorted()`

Finds the position where a value should be inserted in a **sorted array**.

### Example:

```python
a = np.array([10,20,30,40])

np.searchsorted(a, 25)
# Output: 2
```

### Multiple values:

```python
np.searchsorted(a, [15,35])
# Output: [1 3]
```

### Left vs Right:

```python
a = np.array([10,20,20,30])

np.searchsorted(a, 20)              # 1
np.searchsorted(a, 20, side='right') # 3
```

👉 Array must be sorted

---

## 🔷 3. Using `np.argmax()` and `np.argmin()`

Find index of maximum and minimum value.

```python
a = np.array([10,50,30])

np.argmax(a)  # 1
np.argmin(a)  # 0
```

---

## 🔷 4. Using `np.nonzero()`

Returns indices of non-zero elements.

```python
a = np.array([0,1,2,0,3])

np.nonzero(a)
# Output: (array([1,2,4]),)
```

---

## 🔷 5. Boolean Searching (Filtering)

Returns values based on condition.

```python
a = np.array([10,20,30,40])

a[a > 20]
# Output: [30 40]
```

### Multiple conditions:

```python
a[(a > 10) & (a < 40)]
# Output: [20 30]
```

---

## 🔷 Key Points

* `where()` → returns indices
* `searchsorted()` → returns insertion position
* `argmax()` / `argmin()` → index of max/min
* `nonzero()` → indices of non-zero elements
* Boolean indexing → returns filtered values

---

## 🎯 Conclusion

NumPy searching provides efficient methods to locate elements, find positions, and filter data using conditions. It is essential for data analysis and array manipulation.
