```python
df.dropna()
```

```python
df.head(5)
```

## Remove Rows Only If All Values Are Null

```python
df.dropna(how='all')
```

## Remove Nulls From Specific Columns

```python
df.dropna(subset=['Python Marks'])
```

# Filling Null Values → fillna()

## Fill With Single Value

```python
df.fillna(0)
```

## Fill Different Columns With Different Values

## mean,median,mode

```python
df.fillna({
    'Full Name': 'Unknown',

    'Algorithm Marks': df['Algorithm Marks'].mean(),
    'Data Structure Marks': df['Data Structure Marks'].mean(),
    'Python Marks': df['Python Marks'].median(),
    'Location': 'No City'
})
```

## fill_forward_val

```python
df.fillna(method='ffill')
```

# make_permenant_ChangeIn DataFrame

```python
df.fillna({
    'Full Name': 'Unknown',

    'Algorithm Marks': df['Algorithm Marks'].mean(),
    'Data Structure Marks': df['Data Structure Marks'].mean(),
    'Python Marks': df['Python Marks'].median(),
    'Location': 'No City'
},inplace=True)
```
