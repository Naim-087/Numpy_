## Ufunc creation

To create your own ufunc, you have to define a function, like you do with normal functions in Python, then you add it to your NumPy ufunc library with the frompyfunc() method.

The frompyfunc() method takes the following arguments:

1. function - the name of the function.


2. inputs - the number of input arguments (arrays).

3. outputs - the number of output arrays.


```python
import numpy as np
```

```python
def myadd(x,y):
  return x+y

myadd=np.frompyfunc(myadd,2,1)
print(myadd([1,2,3,4],[5,6,3,2]))
```

## Arithamtic_Operations

```python
# ==============================
# NUMPY UFUNC ARITHMETIC (FULL)
# ==============================

import numpy as np

# ------------------------------
# 1. Create Arrays
# ------------------------------
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a =", a)
print("b =", b)

# ------------------------------
# 2. Addition
# ------------------------------
print("\nAddition:")
print(np.add(a, b))
print(a + b)

# ------------------------------
# 3. Subtraction
# ------------------------------
print("\nSubtraction:")
print(np.subtract(a, b))
print(a - b)

# ------------------------------
# 4. Multiplication
# ------------------------------
print("\nMultiplication:")
print(np.multiply(a, b))
print(a * b)

# ------------------------------
# 5. Division
# ------------------------------
print("\nDivision:")
print(np.divide(a, b))
print(a / b)

# ------------------------------
# 6. Power
# ------------------------------
print("\nPower:")
print(np.power(a, 2))
print(a ** 2)

# ------------------------------
# 7. Modulus
# ------------------------------
print("\nModulus:")
print(np.mod(a, 2))
print(a % 2)

# ------------------------------
# 8. Absolute
# ------------------------------
c = np.array([-1, -2, 3])
print("\nAbsolute:")
print(np.absolute(c))

# ------------------------------
# 9. Reciprocal
# ------------------------------
d = np.array([1.0, 2.0, 4.0])
print("\nReciprocal:")
print(np.reciprocal(d))

# ------------------------------
# 10. Negative
# ------------------------------
print("\nNegative:")
print(np.negative(a))

# ------------------------------
# 11. Floor Division
# ------------------------------
print("\nFloor Division:")
print(np.floor_divide(a, b))

# ------------------------------
# 12. Broadcasting
# ------------------------------
print("\nBroadcasting:")
print(np.add(a, 10))

# ------------------------------
# 13. 2D Array Operations
# ------------------------------
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

print("\n2D Multiplication:")
print(np.multiply(x, y))

# ------------------------------
# 14. Reduce (Sum)
# ------------------------------
print("\nReduce (Sum):")
print(np.add.reduce(a))

# ------------------------------
# 15. Accumulate (Cumulative Sum)
# ------------------------------
print("\nAccumulate:")
print(np.add.accumulate(a))

# ------------------------------
# 16. Outer Product
# ------------------------------
print("\nOuter Product:")
print(np.multiply.outer(a, b))
```

## Rounding_Decimal

```python
# =========================================
# NUMPY ROUNDING (COMPLETE SINGLE FILE)
# =========================================

import numpy as np

# -----------------------------------------
# 1. Create Sample Array
# -----------------------------------------
a = np.array([1.2345, 2.5678, 3.9999, -2.7, -3.5])

print("Original Array:")
print(a)

# -----------------------------------------
# 2. ROUND (Nearest Value)
# -----------------------------------------
print("\n1. ROUND (np.round)")
print("Round to 2 decimal places:")
print(np.round(a, 2))

print("\nBanker's Rounding Example:")
print("np.round(3.5) =", np.round(3.5))
print("np.round(2.5) =", np.round(2.5))

# -----------------------------------------
# 3. FLOOR (Always Down)
# -----------------------------------------
print("\n2. FLOOR (np.floor)")
print(np.floor(a))

# -----------------------------------------
# 4. CEIL (Always Up)
# -----------------------------------------
print("\n3. CEIL (np.ceil)")
print(np.ceil(a))

# -----------------------------------------
# 5. TRUNC (Remove Decimal)
# -----------------------------------------
print("\n4. TRUNC (np.trunc)")
print(np.trunc(a))

# -----------------------------------------
# 6. FIX (Towards Zero)
# -----------------------------------------
print("\n5. FIX (np.fix)")
print(np.fix(a))

# -----------------------------------------
# 7. Comparison Summary
# -----------------------------------------
print("\n6. COMPARISON TABLE")
print("Value\tRound\tFloor\tCeil\tTrunc")

for val in a:
    print(f"{val}\t{np.round(val)}\t{np.floor(val)}\t{np.ceil(val)}\t{np.trunc(val)}")

# -----------------------------------------
# 8. 2D ARRAY ROUNDING
# -----------------------------------------
b = np.array([[1.234, 2.789], [3.456, 4.999]])

print("\n7. 2D ARRAY ROUNDING")
print("Original:")
print(b)

print("\nRounded (2 decimals):")
print(np.round(b, 2))

# -----------------------------------------
# 9. PRACTICE SECTION
# -----------------------------------------
print("\n8. PRACTICE")

c = np.array([10.55, 20.25, 30.75])

print("Original:", c)
print("Round:", np.round(c))
print("Floor:", np.floor(c))
print("Ceil:", np.ceil(c))

# -----------------------------------------
# END
# -----------------------------------------
print("\n--- Done Learning Rounding in NumPy ---")
```
