# Regex Filtering

```python
# ==========================================
# REGEX FILTERING FOR YOUR CSV FILE
# ==========================================



# see dataset
print(df.head())



# ==========================================
# 1. STUDENTS WHOSE NAME STARTS WITH A
# ==========================================

print("\nNames starting with A")
print(
    df[
        df['FullName'].str.contains('^A', na=False)
    ]
)



# ==========================================
# 2. STUDENTS WHOSE NAME STARTS WITH A OR F
# ==========================================

print("\nNames starting with A or F")
print(
    df[
        df['FullName'].str.contains('^[AF]', na=False)
    ]
)



# ==========================================
# 3. STUDENTS WHOSE NAME ENDS WITH n
# ==========================================

print("\nNames ending with n")
print(
    df[
        df['FullName'].str.contains('n$', na=False)
    ]
)



# ==========================================
# 4. NAMES CONTAINING 'im'
# ==========================================

print("\nNames containing 'im'")
print(
    df[
        df['FullName'].str.contains('im', case=False, na=False)
    ]
)



# ==========================================
# 5. STUDENT IDS CONTAINING DIGITS
# ==========================================

print("\nStudent IDs containing digits")
print(
    df[
        df['StudentID'].str.contains(r'\d', na=False)
    ]
)



# ==========================================
# 6. STUDENT IDS STARTING WITH PH
# ==========================================

print("\nStudent IDs starting with PH")
print(
    df[
        df['StudentID'].str.contains('^PH', na=False)
    ]
)



# ==========================================
# 7. LOCATION = DHAKA OR SYLHET
# ==========================================

print("\nStudents from Dhaka or Sylhet")
print(
    df[
        df['Location'].str.contains('Dhaka|Sylhet', na=False)
    ]
)



# ==========================================
# 8. INSTRUCTOR NAME STARTING WITH Mr
# ==========================================

print("\nInstructor starts with Mr")
print(
    df[
        df['Instructor'].str.contains('^Mr', na=False)
    ]
)



# ==========================================
# 9. INSTRUCTOR NAME ENDING WITH Karim
# ==========================================

print("\nInstructor ending with Karim")
print(
    df[
        df['Instructor'].str.contains('Karim$', na=False)
    ]
)



# ==========================================
# 10. COMPLETION STATUS CONTAINING Progress
# ==========================================

print("\nIn Progress students")
print(
    df[
        df['CompletionStatus'].str.contains(
            'Progress',
            case=False,
            na=False
        )
    ]
)



# ==========================================
# 11. NAMES WITH EXACTLY 12 CHARACTERS
# ==========================================

print("\nNames with exactly 12 characters")
print(
    df[
        df['FullName'].str.contains(
            '^............$',
            na=False
        )
    ]
)



# ==========================================
# 12. NAMES CONTAINING SPACE
# ==========================================

print("\nNames containing spaces")
print(
    df[
        df['FullName'].str.contains(r'\s', na=False)
    ]
)



# ==========================================
# 13. NAMES CONTAINING ONLY LETTERS/SPACES
# ==========================================

print("\nNames with only letters and spaces")
print(
    df[
        df['FullName'].str.contains(
            r'^[A-Za-z ]+$',
            na=False
        )
    ]
)



# ==========================================
# 14. LOCATION STARTING WITH C OR D
# ==========================================

print("\nLocations starting with C or D")
print(
    df[
        df['Location'].str.contains(
            '^[CD]',
            na=False
        )
    ]
)



# ==========================================
# 15. NOT COMPLETED STUDENTS
# ==========================================

print("\nStudents NOT completed")
print(
    df[
        ~df['CompletionStatus'].str.contains(
            'Completed',
            na=False
        )
    ]
)



# ==========================================
# 16. MULTIPLE CONDITIONS
# ==========================================

print("\nCompleted students from Dhaka")
print(
    df[
        (df['CompletionStatus'].str.contains(
            'Completed',
            na=False
        )) &
        (df['Location'].str.contains(
            'Dhaka',
            na=False
        ))
    ]
)



# ==========================================
# 17. USING loc WITH REGEX
# ==========================================

print("\nUsing loc")
print(
    df.loc[
        df['FullName'].str.contains(
            '^A',
            na=False
        ),
        ['FullName', 'Location']
    ]
)



# ==========================================
# 18. FILTER ROWS WITH MISSING PYTHON MARKS
# ==========================================

print("\nMissing Python Marks")
print(
    df[
        df['Python Marks'].isnull()
    ]
)



# ==========================================
# 19. FILTER ROWS WITH VALID MARKS
# ==========================================

print("\nValid Python Marks")
print(
    df[
        df['Python Marks'].notnull()
    ]
)



# ==========================================
# 20. REGEX CHEAT SHEET
# ==========================================

"""
^       -> start of string
$       -> end of string
.       -> any character
*       -> zero or more
+       -> one or more
?       -> optional
[]      -> character group
\d      -> digit
\w      -> word character
\s      -> whitespace
|       -> OR
"""
```
