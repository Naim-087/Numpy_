## from_morelist_onedataframe

```python
import pandas as pd

name = ["Naim","Rahim","Karim","Sakib","Rifat",
        "Hasan","Jamil","Tanvir","Arif","Nabil"]

age = [21,22,23,20,24,25,22,21,23,24]

department = ["SWE","CSE","EEE","BBA","ENG",
              "SWE","CSE","EEE","BBA","ENG"]

cgpa = [3.75,3.80,3.60,3.50,3.90,
        3.40,3.70,3.65,3.85,3.55]

df = pd.DataFrame(
    [name, age, department, cgpa],

    index=["Name","Age","Department","CGPA"]
)

print(df.T) #make transpose
```

## from_numpy_array

```python
import numpy as np
data = { 'A': np.array([1, 4, 7]),
         'B': np.array([2, 5, 8]),
         'C': np.array([3, 6, 9]) }
df = pd.DataFrame(data)
print(df)
```
